```python
import matplotlib.pyplot as plt

df['Class'].value_counts().plot(kind='bar')
plt.title("Fraud vs Normal Transactions")
plt.xlabel("Class")
plt.ylabel("Count")
plt.show()
```

```python
df['high_amount'] = (df['Amount'] > df['Amount'].quantile(0.95)).astype(int)
```

```python
# Top 1% of transaction amounts are “large”

# We create a binary column: 1 = large, 0 = normal
```

```python
df['flag_large'].value_counts()
```

```python
df['fraud_score'] = df['flag_large'] + df['high_amount']
```

```python
# 0 → safe

# 1 → maybe suspicious

# 2 → high risk

pd.crosstab(df['fraud_score'], df['Class'], normalize='index')
```

```python
import seaborn as sns

sns.countplot(x='fraud_score', hue='Class', data=df)
plt.title("Fraud Score Distribution vs Actual Fraud")
plt.show()
```

```python
## Fraud Detection Findings

- Top 1% largest transactions are more likely to be fraud
- High transaction amounts are a strong indicator
- Combining rules reduces false positives
- Visualizations show patterns clearly
- Feature engineering is key to detection
```
